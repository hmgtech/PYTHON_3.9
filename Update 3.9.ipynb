## Dictionary Unions
**One of my favorite new features with a sleek syntax. If we have two dictionaries a and b that we need to merge, we now use the union operators.
<br>We have the merge operator |:**

```python
a = {1: 'a', 2: 'b', 3: 'c'}
b = {4: 'd', 5: 'e'}
c = a | b
print(c)
```

**And the update operator |=, which updates the original dictionary:**

```python
a = {1: 'a', 2: 'b', 3: 'c'}
b = {4: 'd', 5: 'e'}
a |= b
print(a)
```

```python
a = {1: 'a', 2: 'b', 3: 'c'}
b = {1: 'd', 5: 'e'}
a |= b
print(a)
```

**If our dictionaries share a common key, the key-value pair in the second dictionary will be used:**

```python
a = {1: 'a', 2: 'b', 3: 'c', 6: 'in both'}
b = {4: 'd', 5: 'e', 6: 'but different'}
print(a | b)
```

<b>Dictionary Update with Iterables</b><br>
Another cool behavior of the |= operator is the ability to update the dictionary with new key-value pairs using an iterable object — like a list or generator:

```python
a = {'a': 'one', 'b': 'two'}
b = ((i, i**2) for i in range(3)) #b{0:0, 1:1, 2:4}
a |= b
print(a)
```

**If we attempt the same with the standard union operator | we will get a TypeError as it will only allow unions between dict types.**

**Type Hinting<br>
Python is dynamically typed, meaning we don’t need to specify datatypes in our code.<br>
This is okay, but sometimes it can be confusing, and suddenly Python’s flexibility becomes more of a nuisance than anything else.<br>
Since 3.5, we could specify types, but it was pretty cumbersome. This update has truly changed that, let’s use an example:**

```python
def add_int(val: int):
    return val+val
add_int("r")
```

**In our add_int function, we clearly want to add the same number to itself (for some mysterious undefined reason). But our editor doesn’t know that, and it is perfectly okay to add two strings together using + — so no warning is given.<br>
What we can now do is specify the expected input type as int. Using this, our editor picks up on the problem immediately.**

## New Parser
**This one is more of an out-of-sight change but has the potential of being one of the most significant changes for the future evolution of Python.<br>
Python currently uses a predominantly LL(1)-based grammar, which in turn can be parsed by a LL(1) parser — which parses code top-down, left-to-right, with a lookahead of just one token.<br>
Now, I have almost no idea of how this works — but I can give you a few of the current issues in Python due to the use of this method:<br>
Python contains non-LL(1) grammar; because of this, some parts of the current grammar use workarounds, creating unnecessary complexity.<br>
LL(1) creates limitations in the Python syntax (without possible workarounds). This issue highlights that the following code simply cannot be implemented using the current parser (raising a SyntaxError):**

```python
with (open("a_really_long_foo") as foo,
      open("a_really_long_bar") as bar):
    pass
```

**LL(1) breaks with left-recursion in the parser. Meaning particular recursive syntax can cause an infinite loop in the parse tree. Guido van Rossum, the creator of Python, explains this here.<br>
All of these factors (and many more that I simply cannot comprehend) have one major impact on Python; they limit the evolution of the language.<br>
The new parser, based on PEG, will allow the Python developers significantly more flexibility — something we will begin to notice from Python 3.10 onwards.**
